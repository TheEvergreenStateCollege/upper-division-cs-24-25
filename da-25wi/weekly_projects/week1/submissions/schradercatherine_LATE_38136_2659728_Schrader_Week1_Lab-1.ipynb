```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_csv("/content/drive/MyDrive/data.csv")
df.head(5)
```

```python
df.tail(5)
```

```python
df.dtypes
```

```python
df.info()
```

```python
df["Engine HP"].mean()
```

```python
df[["highway MPG", "city mpg"]].median()
```

```python
df[["highway MPG", "city mpg"]].describe()
```

```python
df[["Make", "city mpg"]].groupby("Make").mean()
```

```python
df["Engine Cylinders"].value_counts()
```

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

```python
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
df.groupby('Drive Mode').size().plot(kind='barh')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.plot.scatter(x='MPG-C', y='MPG-H', alpha=0.5)
```

```python
df.Transmission.value_counts().plot(kind='pie', ylabel='Count', labeldistance=None)
plt.legend()
plt.show()
```
