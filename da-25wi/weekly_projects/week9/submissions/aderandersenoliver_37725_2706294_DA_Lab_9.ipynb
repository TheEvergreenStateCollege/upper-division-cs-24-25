# Imports and Splitting Data


```python
import sys

assert sys.version_info >= (3, 7)
```

```python
from packaging import version
import sklearn

assert version.parse(sklearn.__version__) >= version.parse("1.0.1")
```

```python
import pandas as pd
df = pd.read_csv('./mnist_784.csv')
df.head(2)
```

```python
X = df.iloc[:, 0:784].values
y = df['class'].astype(str).values
```

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

# No PCA

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators = 100)
```

CLF Training took 57 seconds

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print("ACCURACY OF THE MODEL:", metrics.accuracy_score(y_test, y_pred))
```

# Yes PCA

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train)

train_img = scaler.transform(X_train)
test_img = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(.95)
```

```python
pca.fit(train_img)
```

```python
train_img = pca.transform(train_img)
test_img = pca.transform(test_img)
```

```python
pca_clf = RandomForestClassifier(n_estimators = 100)
```

PCA_CLF training took 4 minutes

```python
pca_clf.fit(train_img, y_train)
```

```python
pca_pred = pca_clf.predict(test_img)
```

Slightly wore accuracy compared to non-PCA classifier

```python
print("ACCURACY OF THE MODEL:", metrics.accuracy_score(y_test, pca_pred))
```
