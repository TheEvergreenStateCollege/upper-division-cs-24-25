```python
import numpy as np
from tensorflow.keras.datasets import mnist
from scipy.ndimage import shift
import pandas as pd
import cv2
```

```python
image = pd.read_csv('mnist_784.csv')
image.head(5)
```

```python
def shift_image(image, dx, dy):
    """Shifts the input image by dx and dy."""
    return shift(image, [dy, dx])
```

```python
#Clean
# Function to preprocess images
def preprocess_image(image):
    # Apply Gaussian Blur for noise reduction
    image = cv2.GaussianBlur(image, (5, 5), 0)
    # Resize image to 28x28 (if needed)
    image = cv2.resize(image, (28, 28))
    # Convert to grayscale (already grayscale in this case)
    image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    # Normalize pixel values to range [0, 1]
    image = image / 255.0
    return image
```

```python
def augment_dataset(images): """Creates four shifted copies (left, right, up, down) for each image and adds them to the dataset."""
augmented_images = []
augmented_images.append(image)  # Include original image
for images in image:
    augmented_images.append(shift_image(image, 1, 0))   # Shift right
    augmented_images.append(shift_image(image, -1, 0))  # Shift left
    augmented_images.append(shift_image(image, 0, 1))   # Shift down
    augmented_images.append(shift_image(image, 0, -1))  # Shift up

return np.array(augmented_images)  
```

```python
# Load MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
augmented_train_images = augment_dataset(train_images)
augmented_train_labels = np.repeat(train_labels, 5)
```
