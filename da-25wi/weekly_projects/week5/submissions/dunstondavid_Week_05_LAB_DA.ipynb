```python
import os
import math
import random
import re
import numpy as np

# Basic Syntax and Data Types
name = input("Enter your name: ")
age = int(input("Enter your age: "))
print(f"Hello {name}, you are {age} years old!")

# Control Flow
def check_number(num):
    if num > 0:
        print("Positive number")
    elif num < 0:
        print("Negative number")
    else:
        print("Zero")
num = int(input("Enter a number: "))
check_number(num)

# Functions
def rectangle_area(length, width):
    return length * width
print("Area of rectangle:", rectangle_area(5, 3))

# Data Structures
students = {"Alice": 85, "Bob": 78, "Charlie": 92, "David": 60}
high_scorers = [name for name, score in students.items() if score > 80]
print("Students who scored above 80:", high_scorers)

# File Handling
filename = "names.txt"
names = ["Alice", "Bob", "Charlie"]
with open(filename, "w") as file:
    for name in names:
        file.write(name + "\n")
with open(filename, "r") as file:
    print("Contents of file:")
    print(file.read())

# Object-Oriented Programming (OOP)
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    def display_info(self):
        print(f"Car Brand: {self.brand}, Year: {self.year}")
my_car = Car("Toyota", 2020)
my_car.display_info()

# Exception Handling
try:
    num1 = int(input("Enter numerator: "))
    num2 = int(input("Enter denominator: "))
    print("Result:", num1 / num2)
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")

# Modules and Packages
print("Square root of 16:", math.sqrt(16))

# Iterators and Generators
def square_numbers(n):
    for i in range(1, n+1):
        yield i ** 2
squares = square_numbers(5)
print("Squares:", list(squares))

# Functional Programming
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print("Squared numbers:", squared_numbers)

# Decorators and Closures
def my_decorator(func):
    def wrapper():
        print("Function executed!")
        func()
    return wrapper
@my_decorator
def say_hello():
    print("Hello!")
say_hello()

# Working with Libraries
print("Random number (1-10):", random.randint(1, 10))

# Regular Expressions
def check_phone_number(text):
    pattern = r"\d{3}-\d{3}-\d{4}"
    if re.search(pattern, text):
        print("Valid phone number found!")
    else:
        print("No valid phone number found.")
check_phone_number("My number is 123-456-7890")

# Data Science & Machine Learning Basics
arr = np.array([10, 20, 30, 40, 50])
print("Mean of array:", np.mean(arr))
```
