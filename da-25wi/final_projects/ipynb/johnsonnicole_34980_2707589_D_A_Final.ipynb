# The Spreadsheet of Data: 10 Years of Wildfire Data for Washington State











This is our improved and final version of a spreadsheet representation of the wildfire data. We worked really hard to find a solution that didnt need an external file which would need to use of excel or google spreadsheets. This is a fully independant spreadsheet which is ran and created within the enviorment.

```python
import pandas as pd
from IPython.display import display  # For Jupyter Notebook or other interactive environments

# Data
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 'Total', '10-year average'],
    'Northeast': [338, 271, 401, 423, 483, 518, 383, 345, 486, 434, 499, 4581, 408.20],
    'Northwest': [37, 36, 60, 52, 32, 72, 55, 79, 64, 76, 63, 626, 56.30],
    'Olympic': [26, 36, 50, 28, 32, 29, 18, 24, 37, 26, 20, 326, 30.60],
    'Pacific Cascade': [91, 133, 148, 156, 174, 179, 127, 127, 148, 134, 103, 1520, 141.70],
    'South Puget Sound': [36, 46, 56, 37, 83, 137, 126, 127, 186, 145, 154, 1133, 97.90],
    'Southeast': [103, 96, 139, 142, 152, 134, 128, 111, 132, 146, 128, 1411, 128.30],
}

# Create DataFrame
df = pd.DataFrame(data)

# Add a new column for row-wise totals
df['Row Total'] = df.iloc[:, 1:].sum(axis=1)

# Style the DataFrame to look like an Excel spreadsheet
styled_df = df.style \
    .set_properties(**{
        'text-align': 'center',  # Center-align all text
        'border': '1px solid black',  # Add borders to cells
    }) \
    .set_table_styles([{
        'selector': 'th',  # Style the header row
        'props': [('background-color', '#f7f7f7'), ('font-weight', 'bold'), ('border', '1px solid black')]
    }]) \
    .format(precision=2)  # Format numbers to 2 decimal places

# Display the styled DataFrame
display(styled_df)
```

# 10 Years of Wildfire Data Represented by Scatter Plots

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'Northeast': [338, 271, 401, 423, 483, 518, 383, 345, 486, 434, 499],
    'Northwest': [37, 36, 60, 52, 32, 72, 55, 79, 64, 76, 63],
    'Olympic': [26, 36, 50, 28, 32, 29, 18, 24, 37, 26, 20],
    'Pacific Cascade': [91, 133, 148, 156, 174, 179, 127, 127, 148, 134, 103],
    'South Puget Sound': [36, 46, 56, 37, 83, 137, 126, 127, 186, 145, 154],
    'Southeast': [103, 96, 139, 142, 152, 134, 128, 111, 132, 146, 128],
    'Total': [631, 618, 854, 838, 956, 1069, 837, 813, 1053, 961, 967]
}

# Create DataFrame
df = pd.DataFrame(data)

# Regions to plot
regions = ['Northeast', 'Northwest', 'Olympic', 'Pacific Cascade', 'South Puget Sound', 'Southeast']

# Plotting
plt.figure(figsize=(15, 10))
for i, region in enumerate(regions, 1):
    plt.subplot(2, 3, i)
    plt.scatter(df['Year'], df[region])
    plt.title(region)
    plt.xlabel('Year')
    plt.ylabel('Number of Fires')
    plt.grid(True)

plt.tight_layout()
plt.show()
```
