# Plotting demos

Python has different plotting libraries.  Here are some common ones:
* Matplotlib
* Seaborn 
* Plotly
* Altair
* Bokeh

Part of your job as a data-person (analyst, scientist) is to become familiar with one or more of these libaries and get proficient at using them to make plots that advance a narrative. 

This demo uses bits and pieces of matplotlib and seaborn.  The interested reader is encouraged to google around for the plotting demos and to experiment wildly.

```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris = load_iris(as_frame=True)
df = iris.data
df['variety'] = iris['target'].map(dict(zip([0, 1, 2], iris['target_names'])))
```

```python
df.head()
```

```python
# rename columns for easy access
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'variety']
```

```python
# scatterplot
sns.scatterplot(x=df.index, y=df.sepal_width, hue=df.variety)
```

```python
sns.histplot(df.sepal_width)
```
